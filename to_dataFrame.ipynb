```python
import pandas as pd
import os

image_folder = 'D:\\minkwan\\졸업작품\\coordikitty-ML-DL\\image_root_dir'
label_folder = 'D:\\minkwan\\졸업작품\\coordikitty-ML-DL\\label_dir'

image_files = os.listdir(image_folder)
label_files = os.listdir(label_folder)
```

```python
# 라벨링 파일을 사전으로 변환하여 빠른 검색 가능
label_dict = {}
for label_file in label_files:
    label_parts = label_file.split('_')
    label_number = label_parts[0]  # 첫 번째 부분 (예: 0)
    label_parts[4] = label_parts[4].replace(".json", "")
    label_label = '_'.join(label_parts[2:])  # 라벨 부분 (예: BOTTOMS_LONG_COTTON)

    # 중복된 라벨 번호가 있을 경우 리스트로 저장
    if label_number not in label_dict:
        label_dict[label_number] = []
    label_dict[label_number].append((label_label, label_file))  # 라벨과 원본 파일 이름 저장
```

```python
# 매칭된 결과를 저장할 리스트
matched_data = []
```

```python
# 이미지 파일과 라벨링 파일 매칭
for image_file in image_files:
    # 이미지 파일 이름에서 필요한 부분 추출
    image_parts = image_file.split('_')
    image_number = image_parts[-2]  # 마지막 부분 (예: 6)
    image_label = '_'.join(image_parts[0:3])  # 라벨 부분 (예: BOTTOMS_LONG_COTTON)

    # 라벨링 파일 찾기
    if image_number in label_dict:
        for label_label, label_file in label_dict[image_number]:
            if image_label == label_label:
                # 라벨링 파일의 원본 이름 추가
                matched_data.append({
                    'image_file': image_file,
                    'label_file': label_label,
                    'original_label_file': label_file  # 라벨링 파일 원본 이름 추가
                })
```

```python
# 데이터프레임 생성
df = pd.DataFrame(matched_data)
```

```python
df = df.drop(columns=['label_file'], errors='ignore')
```

### json파일의 컬럼을 데이터프레임 컬럼으로 추가

```python
import json
```

```python
label_folder = 'D:\minkwan\졸업작품\coordikitty-ML-DL\label_dir'
```

```python
for index, row in df_res.iterrows():
    label_file = row['label_file']
    file_path = os.path.join(label_folder, label_file)
    
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 필요한 정보 추출하여 새로운 컬럼 추가
    df_res.at[index, 'large_category'] = data.get('large_category')
    df_res.at[index, 'medium_category'] = data.get('medium_category')
    df_res.at[index, 'small_category'] = data.get('small_category')
    df_res.at[index, '핏'] = ', '.join(data.get('핏', []))  # 리스트를 문자열로 변환
    df_res.at[index, '촉감'] = ', '.join(data.get('촉감', []))
    df_res.at[index, '신축성'] = ', '.join(data.get('신축성', []))
    df_res.at[index, '비침'] = ', '.join(data.get('비침', []))
    df_res.at[index, '두께'] = ', '.join(data.get('두께', []))
    df_res.at[index, '계절'] = ', '.join(data.get('계절', []))
    df_res.at[index, 'major_style'] = data.get('major_style')
    df_res.at[index, 'minor_style'] = data.get('minor_style')
```
